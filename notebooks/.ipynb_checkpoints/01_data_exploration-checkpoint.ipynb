```python
# Notebook 01 - Eksplorasi Data: Konsumsi Listrik Rumah Tangga

# %% [markdown]
# # Notebook 01: Eksplorasi Data Awal
# 
# Notebook ini bertujuan untuk melakukan eksplorasi awal terhadap dataset konsumsi listrik rumah tangga berdasarkan aktivitas harian.

# %%
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set style default
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# %% [markdown]
# ## 1. Memuat Dataset

# %%
df = pd.read_csv("../data/generated/listrik_rumah.csv")
print(f"Jumlah baris: {df.shape[0]}")
print(f"Jumlah kolom: {df.shape[1]}")
df.head()

# %% [markdown]
# ## 2. Informasi Umum Dataset

# %%
df.info()

# %% [markdown]
# ## 3. Statistik Deskriptif

# %%
df.describe()

# %% [markdown]
# ## 4. Cek Data Kosong

# %%
df.isnull().sum()

# %% [markdown]
# ## 5. Visualisasi Distribusi Target

# %%
sns.histplot(df['konsumsi_listrik_kwh'], kde=True)
plt.title("Distribusi Konsumsi Listrik (kWh)")
plt.xlabel("kWh")
plt.ylabel("Frekuensi")
plt.show()

# %% [markdown]
# ## 6. Korelasi Antar Variabel Numerik

# %%
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriks Korelasi")
plt.tight_layout()
plt.show()

# %% [markdown]
# ## 7. Korelasi terhadap Target

# %%
if 'konsumsi_listrik_kwh' in corr.columns:
    target_corr = corr['konsumsi_listrik_kwh'].drop('konsumsi_listrik_kwh')
    target_corr.sort_values(ascending=False).plot(kind='bar')
    plt.title("Korelasi Fitur dengan Konsumsi Listrik")
    plt.ylabel("Koefisien Korelasi")
    plt.show()

# %% [markdown]
# ## 8. Visualisasi Fitur Kategorikal (jika ada)

# %%
categorical_cols = df.select_dtypes(include='object').columns.tolist()
for col in categorical_cols:
    sns.boxplot(x=col, y='konsumsi_listrik_kwh', data=df)
    plt.title(f"Distribusi Konsumsi Listrik berdasarkan {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# %% [markdown]
# ## 9. Simpan Laporan Eksplorasi (Opsional)

# %%
with open("../reports/final_report.md", "a") as f:
    f.write("\n\n## Statistik Deskriptif\n")
    f.write(df.describe().to_markdown())
    f.write("\n\n")
    f.write("## Korelasi dengan Target\n")
    f.write(target_corr.sort_values(ascending=False).to_markdown())
```
